### Follow the instructions to recreate the plots


```python
# Use this data for the exercises
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook.  
If you aren't using the jupyter notebook you need to use plt.show() to display a figure**

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Exercise 1

- **Create a figure object called fig using plt.figure()**
-**Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.**
-**Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x)
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
#2
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
#2
ax1.plot(x,y,'r')
ax1.set_ylabel('y')
ax1.set_xlabel('x')

ax2.plot(x,y,'r')
ax2.set_ylabel('y')
ax2.set_xlabel('x')
fig
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
ax1.plot(x,z)
ax1.set_ylabel('Z')
ax1.set_xlabel('X')



ax2.plot(x,y)
ax2.set_ylim([30,50])
ax2.set_xlim([20,22])
ax2.set_title('zoom')
ax2.set_ylabel('Y')
ax2.set_xlabel('x')
fig
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
axes[0].plot(x,y,'b--',linewidth=3)
axes[1].plot(x,z,'r',linewidth=3)
fig
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=[12,2])
axes[0].plot(x,y,'r',linewidth=3)
axes[1].plot(x,z,'b--',linewidth=3)
```
